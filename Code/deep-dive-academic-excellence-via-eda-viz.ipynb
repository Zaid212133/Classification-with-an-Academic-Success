```python
def bar_char(column):
  bar_column = (data_train[column].value_counts(normalize=True) * 100).sort_values(ascending=False)
  fig = px.bar(
      bar_column,
      x=bar_column.index,
      y=bar_column.values,
      color=bar_column.index.astype(str),
      title=f'{column}',
      labels={'x': f"{column}" , 'y': 'Counts'},
      text=bar_column.apply(lambda x : f"{x:.2f}%"),
      template='plotly_dark'
  )

  fig.update_traces(
      textposition="outside",
      textfont={"size": 12, "family": "consolas"}
  )

  fig.update_layout(
      xaxis=dict(
          tickmode='linear',
          dtick=1
      )
  )

  return iplot(fig)
```

```python
def Outliers_percentage (data):
  columns_int=int_col
  percentage_out_liers=[]
  for column in columns_int:
    Q1=data[column].quantile(.25)
    Q3=data[column].quantile(.75)
    IQR=Q3-Q1
    lower_bounds=Q1-IQR*1.5
    upper_bounds=Q3+IQR*1.5
    out_liers=data[(data[column]<lower_bounds)| (data[column]>upper_bounds)]
    percentage_=len(out_liers)/len(data)*100
    percentage_out_liers.append(percentage_)
  plt.figure(figsize=(10,10))
  sns.barplot(x=columns_int,y=percentage_out_liers,palette='Set3')
  for i,v in enumerate(percentage_out_liers):
    plt.text(i,v+.005,f"{v:.2f}%",ha='center',va='bottom',fontsize=7)
  plt.title('Percentage of Outliers in Each Column')
  plt.xlabel('Columns')
  plt.ylabel('Percentage of Outliers')
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
#data_train
Outliers_percentage(data_train)
```

```python
#data_test
Outliers_percentage(data_test)
```

```python
def handling_out_liers(data):
    columns=int_col
    for column in columns:
      Q1=data[column].quantile(.25)
      Q3=data[column].quantile(.75)
      IQR=Q3-Q1
      lower_bounds=Q1-IQR*1.5
      upper_bounds=Q3+IQR*1.5
      data[column]=np.where(data[column]<lower_bounds,lower_bounds,data[column])
      data[column]=np.where(data[column]>upper_bounds,upper_bounds,data[column])
    return data


data_train=handling_out_liers(data_train)
data_test=handling_out_liers(data_test)
```

```python
accuracys_models=dict()
def evulation_model(model,X_train,y_train,X_test,y_test,model_name):
  model_classifier=model
  model_classifier.fit(X_train_stand_data_frame,y_train)
  y_predict_train=model_classifier.predict(X_train_stand_data_frame)
  y_predict_test=model_classifier.predict(X_test_stand_data_frame)
  y_proba=model_classifier.predict_proba(X_test_stand_data_frame)
  Auc=roc_auc_score(y_test,y_proba,multi_class='ovr').round(2)
  accuracy_train=accuracy_score(y_train,y_predict_train).round(2)
  accuracy_test=accuracy_score(y_test,y_predict_test).round(2)
  print(f'The accuracy data train : {accuracy_train}'.title())
  print("*"*len(f'The accuracy data train : {accuracy_train}'))
  print(f'The accuracy data test : {accuracy_test}'.title())
  print("*"*len(f'The accuracy data train : {accuracy_train}'))
  print(f"The roc curv score : {Auc}".title())
  print("*"*len(f'The accuracy data train : {accuracy_train}'))
  print(classification_report(y_predict_train,y_train))
  print("*"*len(f'The accuracy data train : {accuracy_train}'))
  print(classification_report(y_predict_test,y_test))
  confusion_matrix_train=confusion_matrix(y_predict_train,y_train)
  confusion_matrix_test=confusion_matrix(y_predict_test,y_test)
  plt.figure(figsize=(10,5))
  plt.subplot(1,2,1)
  sns.heatmap(confusion_matrix_train,annot=True,fmt='.1f',linewidths=2, linecolor='lightgrey')
  plt.title('confusion_matrix data train'.title())
  plt.subplot(1,2,2)
  sns.heatmap(confusion_matrix_test,annot=True,fmt='.1f',linewidths=2, linecolor='lightgrey')
  plt.title('confusion_matrix data test'.title())
  plt.show()
  accuracys_models[model_name] = {'TEST ACCURACY':accuracy_test,'TRAIN ACCURACY':accuracy_train,'AUC':Auc}


```
